```python
import pandas as pd

def rearrange_products_table(products: pd.DataFrame) -> pd.DataFrame:
    return products.melt(id_vars=['product_id'], value_vars=['store1', 'store2', 'store3'], var_name='store', value_name='price').dropna()
    # sort_values(by=['product_id', 'store']).reset_index(drop=True)
    # 后面这两个步骤无关紧要, 没说要排序
```

```python
data = [[0, 95, 100, 105], [1, 70, None, 80]]
Products = pd.DataFrame(data, columns=['product_id', 'store1', 'store2', 'store3']).astype({'product_id':'int64', 'store1':'int64', 'store2':'int64', 'store3':'int64'}, errors="ignore")
Products
```

```python
rearrange_products_table(Products)
```
