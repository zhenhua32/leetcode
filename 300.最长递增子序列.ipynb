```python
"""
记忆化搜索

这个主要是要化简问题. 

1. 先把它定义成特殊形式, 就是一个数组的最后一个值看作是不变的, 
就是必须要用上这最后一个元素. 那么就是求所有比它小的元素的最长子序列,
然后求它们的最大值, 再加上 1 就是当前的答案.

1. 学会最后怎么合并结果, 也就是数组增长一个元素后应该怎么求解
假设以前的结果是 A, 加了一个元素后结果是 B, max(A, B) 就是最终的结果

"""

from typing import List

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        # 存储以前解过的子问题
        result = dict()

        def my_func(i):
            """
            返回子数组的长度, 以及当前子数组
            """
            if i in result:
                return result[i]

            res = 0
            for j in range(i):
                if nums[j] < nums[i]:
                    res = max(res, my_func(j))
            
            result[i] = res + 1
            return res + 1

        # 这个是求了所有的解, 然后取最大值
        return max([my_func(i) for i in range(len(nums))])
```

```python
"""
递推
"""

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        f = [0] * len(nums)
        for i, x in enumerate(nums):
            for j, y in enumerate(nums[:i]):
                if x > y:
                    # 当前的结果和之前的最好结果比较, 取最大值
                    f[i] = max(f[i], f[j])
            f[i] += 1
        return max(f)
```

```python
"""
二分查找法的思想
"""
from bisect import bisect_left

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        g = []
        for x in nums:
            j = bisect_left(g, x)
            if j == len(g):  # >=x 的 g[j] 不存在
                g.append(x)
            else:
                g[j] = x
        return len(g)
```

> 这道题有两种做法，一种是DP也就是动态规划，很简单，第i个元素之前的最小上升子序列的长度无非就是max(dp[i],dp[j]+1),那么另一种做法就是二分查找法，也很简单，无非就是再新建一个数组，然后第一个数先放进去，然后第二个数和第一个数比较，如果说大于第一个数，那么就接在他后面，如果小于第一个数，那么就替换，一般的，如果有i个数，那么每进来一个新的数，都要用二分查找法来得知要替换在哪个位置的数。因为有个for循环，所以是O(N),在加上循环里有个二分查找，所以最后是O(NlogN)的时间复杂度。
