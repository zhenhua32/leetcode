```python
class Solution:
    cache_result = {}

    def jump(self, nums: List[int]) -> int:
        key = tuple(nums)
        if key in self.cache_result:
            return self.cache_result[key]

        len_nums= len(nums)
        if len_nums == 2:
            return 1
        if len_nums <= 1:
            return 0

        # 找到所有可能一步跳到终点的索引位置
        index = []
        for i, x in enumerate(nums[:-1]):
            if i + x >= len_nums - 1:
                index.append(i)
        
        # print(index)
        # 如果直接跳到 0 了, 返回 1
        if 0 in index:
            self.cache_result[key] = 1
            return 1

        # 用每个索引去尝试, 找次数最少的. 因为这里可能会情况为 0 的, 所以替换成 1
        # 然后加 1 次, 当前尝试的最后一步
        result =  1 + (min([self.jump(nums[:x+1]) for x in index]) or 1)
        self.cache_result[key] = result
        return result
```
