```python
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        slen = len(s)
        if slen <= 1:
            return 0
        
        max_len = 0
        stack = 0
        stack_map = {}
        last_stack = -1
        for x in s:
            if x == "(":
                # 每次遇到, 栈里就会多一个数据, 保存上一个栈的状态
                last_stack = stack
                stack += 1
            else:
                if stack > 0:
                    stack -= 1
                    # 先加
                    stack_map[stack] = stack_map.get(stack, 0) + 2
                    # 如果等于上一个栈状态, 直接累加. 比如上一个是 3, 然后经过中间一段(3 以上的)后, 又回到 3 了
                    if stack + 1 == last_stack:
                        stack_map[stack] += stack_map[last_stack]
                        # 重置为 0
                        stack_map[last_stack] = 0
                        # 又多匹配了一个, 所以上一次栈状态要减少 1 格
                        last_stack -= 1
                    max_len = max(max_len, stack_map[stack])
                else:
                    # 没有找到, 野生的 ) 就重置掉
                    stack = 0
                    stack_map = {}
        return max_len
```
